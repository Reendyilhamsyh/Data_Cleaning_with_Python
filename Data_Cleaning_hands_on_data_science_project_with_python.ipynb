2.Membuat if else logic untuk mengganti data kategori menjadi numerik:

```python
#2.Membuat if else logic untuk mengganti data kategori menjadi numerik:
class_list = []#list atau array(kalo pake np baru array sungguhan) kosong untuk menampung
for x in titanic['class']:#untuk setiap data pada titanic column class
    if x == 'First':
        class_list.append(1)
    elif x == 'Second':
        class_list.append(2)
    else :
        class_list.append(3)
```

Hands On Missing Value
Delete the missing value

```python
#Hands On Missing Value
#Delete the missing value
#Preprocessing (drop data yang tidak digunakan dan lakukan encoding)
#drop column column yang tidak digunakan terlebih dahulu
titanic = sns.load_dataset("titanic")
titanic.drop(['who','adult_male','alive','alone','embark_town'],axis = 'columns', inplace=True)
titanic.columns = ['Survived','Pclass','Sex','Age','SibSp','Parch','Fare','Embarked','Class','Deck']
```

```python
titanic['Sex'] = [1 if x == 'male' else 0 for x in titanic['Sex']]
titanic['Class'] = [1 if x == 'First' else (2 if x == 'Second' else 3) for x in titanic['Class']]
```

```python
titanic.head()
```

Cek info datanya dengan method .info()

```python
#Cek info datanya dengan method .info()
titanic.info()
#Dalam info dibawah,dapat diartikan bahwa total jumlah baris adalah 891,pada kolom Age,data yang
#Non-Null (terisi) adalah 714 yang berarti dalam kolom age terdapat missing value,berlaku juga pada data Deck yang hanya terisi 203
```

Method Drop the column

```python
#Method Drop the column
#Jika kolom tersebut memiliki banyak missing values (lebih dari 50-60%),maka sulit rasanya bisa mendapatkan
#metode imputasi yang akurat.apalagi,jika kolom tersebut dirasa memiliki informasi yang minim.
#jika hal ini terjadi,maka hal yang tidak dapat dilakukan adalah menghilangkan seluruh kolom
#(tidak menggunakan kolom tersebut dalam EDA/Modelling).Hal ini yang akan kita lakukan pada kolom "Deck"
titanic.drop("Deck",axis='columns',inplace=True)
titanic.columns
#Kolom deck telah hilang dari dataframe kita.kode inplace = true,berarti kolom deck di drop permanen
#Jika kita tidak ingin men-drop secara permanen maka tidak perlu diberikan inplace=True
```

Method Remove the rows

```python
#Method Remove the rows
#apabila missing values kurang dari 50%,maka hal yang dapat dilakukan adalah menghilangkan semua baris dengan
#missing values.ini adalah metode yang paling mudah,namun memiliki beberapa konsekuensi:
#-Kehilangan informasi dari dalam kolom lain yang penting(dalam kebanyakan kasus,baris yang di-
#hilangkan itu sebenarnya tidak bermasalah di kolom lain)
#-Dataset size menjadi sedikit,bila yang dihilangkan cukup banyak
#Method ini dapat dilakukan pada kolom embarked,karena kolom ini hanya 2 yang missing value.
#Cara 1 Menagmbil baris-baris yang embarked nya tidak ada missing value,menggunakan method .notna()
#df_update = titanic.copy()
#df_update = df_update[df_update['Embarked'].notna()].reset_index(drop = True)

#Cara 2,menggunakan negasi,memilih yang kosong(null) dan di negasikan menggunakan method .isnull()
df_update = titanic.copy()
df_update = titanic[~df_update['Embarked'].isnull()].reset_index(drop = True)
df_update.info()
```

Mengisi dengan mean/median,dapat menggunakan metode .fillna

```python
#Mengisi dengan mean/median,dapat menggunakan metode .fillna:
#Membuat kolom age yang baru dengan cara(mengisi nilai median):
df_update = titanic.copy()
df_update['Age']
df_update['Age'].fillna(df_update['Age'].median())
#-Ambil kolom age yang lama(terdapat missing value)
#-isi missing value dengan nilai median dari kolom age
#-Mengakibatkan
```

presentasikan dengan histogram plot

```python
#kita presentasikan dengan histogram plot
sns.histplot(data = df_update, x = 'Age')
```

membuat kolom age yang baru dengan cara (mengisi nilai modus)

```python
#membuat kolom age yang baru dengan cara (mengisi nilai modus):
#Salah satu dampak buruk dari mengisi missing value dengan mean/medain adalah terjadinya
#distribusi data yang cukup abnormal dan tidak natural
df_update['Embarked']
df_update['Embarked'].fillna(df_update['Embarked'].mode()[0])
df_update.info()
```

Implementasi machine learning menggunakan knn

```python
#menggunakan KNN
#Gunakan library KNNImputer
from sklearn.impute import KNNImputer
```

```python
imputer = KNNImputer()
```

Jika ingin menggunakan machine learning,maka datanya harus berupa angka terlebih dahulu semuanya/ di encode

```python
#Jika ingin menggunakan machine learning,maka datanya harus berupa angka terlebih dahulu semuanya(Encode)
#dalam contoh,kolom embarked masih berupa data string,maka perlu di encode
df_update = titanic.copy()
df_update = df_update.join(pd.get_dummies(df_update['Embarked'], prefix='Embarked'))
df_update.drop('Embarked',axis='columns',inplace=True)
```

gunakan algoritma machine learning KNN,dengan menggunakan metode fit_transform

```python
#gunakan algoritma machine learning KNN,dengan menggunakan metode fit_transform(tabel)
df_update = imputer.fit_transform(df_update)
```

```python
#masukkan nilai update ke dalam kolom:
df_update = pd.DataFrame(df_update)
# Get the original column names from 'titanic' before imputation
# Instead of using the original column names, generate new column names for the one-hot encoded and imputed data
num_cols = df_update.shape[1]  # Get the number of columns in df_update
colname = [f'col_{i}' for i in range(num_cols)] # Generate new column names
df_update.columns = colname # Assign the new column names
df_update
```

Hands on outlier

```python
#1.Cek apakah ada outlier (outlier identification)
#1.menggunakan boxplot
# Load the Boston housing data into a pandas DataFrame called 'boston_df'
boston_df = pd.read_csv("HousingData.csv")
boston_df.head()
```

```python
#1.Cek apakah ada outlier (outlier identification)
#1.menggunakan boxplot
sns.boxplot(x=boston_df['DIS'])
plt.show()
#Berarti hasil di bawah terdapat beberapa outlier,yaitu yang "DIS" nya > 10
```

```python
#Membuat kolom baru,untuk mengecek mana saja baris yang termasuk outlier
# Create the 'outlier_DIS' column and mark outliers
boston_df['outlier_DIS'] = np.where(boston_df['DIS'] > 10, 'outlier', 'not_outlier') # Create 'outlier_DIS' column

#Membuat kolom baru,untuk mengecek mana saja baris yang termasuk outlier
boston_df[boston_df['outlier_DIS'] == 'outlier']
```

```python
#selain menggunakan boxplot,dapat menggunakan scatterplot
#namun perlu di ingat bahwa untuk mendapatkan outlier,harus involving 2 numerical column,kenapa?
#karena terkadang univariate outlier analysis tidak dapat menemukan outlier
#contoh:
sns.boxplot(x=boston_df['INDUS'])
plt.show()
```

```python
sns.boxplot(x=boston_df['TAX'])
plt.show()
```

```python
#dua boxplot di atas bila di lihat masing masing tidak terdapat outlier,namun jika kita hubungkan
#2 column di atas menggunakan hasilnya
fig, ax = plt.subplots(figsize=(16, 8))
ax.scatter(boston_df['INDUS'], boston_df['TAX'])
ax.set_xlabel('INDUS - Proportion of non retail business acres per town')
ax.set_ylabel('TAX - Full value property tax rate per $10.000')
plt.show()
#dapat dilihat ternyata ada beberapa daya yang berada di posisi yang "janggal" apabila dilihat dari
#scatterplot.Hal ini balik lagi ke pemahaman data scientist / data analyst.karena kalau menggunakan
#scatterplot,tidak ada definisi 'IQR' statistik yang menentukan objek tersebut outlier atau bukan
#Berarti dalam kasus dibawah ,jika sebuah row memenuhi salah satu poin berikut:
#-INDUS yang > 16,atau
#-TAX yang >500
#maka mereka adalah outlier
```
